# API Call

```python
import requests

def get_prayer_times(date, latitude, longitude,day):
    api_key = 'http://api.aladhan.com/v1/calendarByAddress/:year/:month'
    #url = f'https://api.aladhan.com/v1/timingsByCoordinates/{date}?latitude={latitude}&longitude={longitude}&method=2'  # 'method=2' corresponds to the Islamic Society of North America (ISNA) calculation method.
    url = f'https://api.aladhan.com/v1/calendar/{date}?latitude={latitude}&longitude={longitude}'
#     headers = {
#         'Authorization': f'Bearer {api_key}',
#     }
    print(url)
    response = requests.get(url)#, headers=headers)
    if response.status_code == 200:
        data = response.json()
        prayer_times = data['data'][day]['timings']
        return prayer_times
    else:
        print(f"Error fetching prayer times: {response.status_code} - {response.text}")
        return None


#Example to get prayer Timmings from Islamabad     
if __name__ == '__main__':
    date = '2017/4'  # Replace this with your desired date in YYYY-MM-DD format.
    #islamabad 33.738045, 73.084488
    latitude = 33.738045  # Replace this with your current latitude.
    longitude = 73.084488  # Replace this with your current longitude.
    day = 10

    prayer_times = get_prayer_times(date, latitude, longitude,day)
    if prayer_times:
        print(prayer_times)
    else:
        print("Failed to get prayer times.")
```

# New Code

```python
N = 3 #increment this value for new person record
post_no = 1
prv_current_rakah = None
prv_posture_instance = None

QayamType = PostureType("QayamType")
RukuType = PostureType("RukuType")
QoumType = PostureType("QoumType")
SajdaType = PostureType("SajdaType")
JalsaType = PostureType("JalsaType")
TashahudType = PostureType("TashahudType")

for idx,(prson, time, pstr, rkh, nmz, cr_rkh, sl_unit, personLocation) in enumerate(zip(data['person_id'],data['Timestamp'],data['Activity'],data['Entered_Rakah'],data['Namaz'],data['Curr_rakah'],data['Salah Unit'],data['location'])):
    
    person = Person(prson)
    salah_instance = Salah('S' + str(N))
    person.performs = [salah_instance]
    salah_instance.isPerformedBy = [person]
    
    #Date processing:
    ym,d = date_processing(time)

    ptimes = get_prayer_times(ym, latitude, longitude,d)
    
    #adding location
    print(type(personLocation),' ',personLocation)
    SalahLocation_instance = SalahLocation(personLocation)
    salah_instance.location = [SalahLocation_instance]

    
    
    if idx==0:
        salah_instance.isStartedAtTime = [time]
    
    if idx+1 == len(data):
        salah_instance.isCompletedAtTime = [time]
    
    if sl_unit=='Sunnah':
        unit_name = 'SN'
        sala_unit = SalahUnit("S"+str(N)+ '_' + unit_name+'_Salah')
        sala_unit_category = SalahUnitCategory('Sunnah')
        sala_unit.hasSalahUnitType = [sala_unit_category]
    else:
        unit_name = 'FZ'
        sala_unit = SalahUnit("S"+str(N)+ '_' + unit_name+'_Salah')
        sala_unit_category = SalahUnitCategory('Farz')
        sala_unit.hasSalahUnitType = [sala_unit_category]
    
    salah_instance.hasSalahUnit = [sala_unit]
    
    if nmz == 'Isha':
        salah_instance.hasSalahType = [onto.Isha]
        salah_instance.hasWindowStartTime = [ptimes['Isha']]
        salah_instance.hasWindowEndTime = [ptimes['Fajr']]
        
    elif nmz == 'Maghrib':
        salah_instance.hasSalahType = [onto.Maghrib]
        salah_instance.hasWindowStartTime = [ptimes['Maghrib']]
        salah_instance.hasWindowEndTime = [ptimes['Isha']]

    elif nmz == 'Asar':
        salah_instance.hasSalahType = [onto.Asar]
        salah_instance.hasWindowStartTime = [ptimes['Asr']]
        salah_instance.hasWindowEndTime = [ptimes['Maghrib']]
        
    elif nmz == 'Zuhr':
        salah_instance.hasSalahType = [onto.Zuhr]
        salah_instance.hasWindowStartTime = [ptimes['Dhuhr']]
        salah_instance.hasWindowEndTime = [ptimes['Asr']]

        
    elif nmz == 'Fajar':
        salah_instance.hasSalahType = [onto.Fajar]
        salah_instance.hasWindowStartTime = [ptimes['Fajr']]
        salah_instance.hasWindowEndTime = [ptimes['Sunrise']]



    current_rakah = Rakah("S"+str(N)+ '_' + unit_name+'_Salah'+'_Rakah'+str(cr_rkh))
#     current_rakah.belongsToUnit = [sala_unit]
    
    if prv_current_rakah != None:
        prv_current_rakah.rakahFollowedByRakah = [current_rakah]
    
        sala_unit.containsRakah.append(current_rakah)
    
    if prv_current_rakah != current_rakah:
        post_no = 1
    
    prv_current_rakah = current_rakah
    
    
    ## Making the Postures
    posture_instance = Postures("S"+str(N)+ '_' + unit_name+'_Rakah'+str(cr_rkh) + '_P'+str(post_no))
    
    current_rakah.containsPosture.append(posture_instance)
    
    if prv_posture_instance != None:
        prv_posture_instance.followedBy = [posture_instance]
        prv_posture_instance.postureEndsAtTime = [time]
        
    posture_instance.postureStartsAtTime = [time]
    
    
    if post_no == 1:
        current_rakah.hasStartingPosture = [posture_instance]
        
    if pstr == 'Qayam':
        posture_instance.hasPostureType = [QayamType]
    elif pstr == 'Ruku':
        posture_instance.hasPostureType = [RukuType]
    elif pstr == 'Qoum':
        posture_instance.hasPostureType = [QoumType]
    elif pstr == 'Sajda1' or pstr == 'Sajda2':
        posture_instance.hasPostureType = [SajdaType]
    elif pstr == 'Jalsa1':
        posture_instance.hasPostureType = [JalsaType]
    elif pstr == 'Tashahud':
        posture_instance.hasPostureType = [TashahudType]
    
    
    prv_posture_instance = posture_instance
    post_no+=1
```
